```python
import gmaps
import pandas as pd

# Google developer API key
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Import CSV files

Seattle = pd.read_csv('../Resources/sea_listings.csv')
Seattle.dropna
Boston = pd.read_csv('../Resources/bos_listings.csv')
Boston.dropna
Chicago = pd.read_csv('../Resources/chi_listings.csv')
Chicago.dropna
San_Diego = pd.read_csv('../Resources/sd_listings.csv')
San_Diego.dropna
Austin = pd.read_csv('../Resources/aus_listings.csv')
Austin.dropna
```

```python
# Merge CSV files
Combined = pd.concat([Seattle, Boston, Chicago, San_Diego, Austin], axis=0)

# Index by location
Combined.set_index('market', inplace=True)
Combined.sort_index(inplace=True, ascending=False)

# Print to CSV
Combined.to_dense().to_csv('../Resources/combined_data.csv', index = False, sep=',', encoding='utf-8')

# View Combined DataFrame
Combined.head()
```
